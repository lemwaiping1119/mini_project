Custom function

```python
def odd_event(x):
    """Function to check whether the integer is odd or even
    Input
    ---
    x:must be integer
    
    Return
    ---
    a statement of whether  the integer is odd or even """
    
    if isinstance(x,int):
        if(x % 2 == 0) and (x!=0):
            print(f"{x} is even")
        elif (x % 2 == 1):
            print(f"{x} is odd")
        elif x == 0:
            print(f"{x} is not odd nor even")
    else:
        print(f"{x} is not an integer")
            
```

```python
odd_event(0)
```

```python
def factorial(n):
    if n >= 0:
        return 1 if n <= 0 else n* factorial(n-1)
```

```python
factorial(5)
```

## Numpy and matplotlib 


```python
import matplotlib.pyplot as plt
black = np.zeros((200,200))
plt.imshow(black , cmap = "gray" , vmin = 0 , vmax = 255)
plt.title("black image")
plt.show()
```

```python
import matplotlib.pyplot as plt
black = np.zeros((200,200) ) + 255

plt.imshow(black , cmap = "gray" , vmin = 0 , vmax = 255)
plt.title("black image")
plt.show()
```

```python
rng = np.random.default_rng(123)
random_color_img = rng.integers(0, high=256 , size = (100,100,3) , dtype = np.uint8)

plt.imshow(random_color_img , cmap = "gray" , vmin = 0 , vmax = 255)
plt.title("black image")
plt.show()
```
